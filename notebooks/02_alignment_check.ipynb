# 02 — Validación de Alineación

Validación visual de los pares alineados OCR ↔ ground truth:
- Muestras aleatorias de pares
- Visualización de diferencias (diff)
- Distribución de scores de similitud

```python
import json

with open('../data/pairs/dataset.json', 'r') as f:
    pairs = json.load(f)

# Mostrar primeros 5 pares
for i, pair in enumerate(pairs[:5]):
    print(f'=== Par {i+1} (similitud: {pair["similarity"]}) ===')
    print(f'OCR:    {pair["ocr"][:100]}...')
    print(f'GT:     {pair["ground_truth"][:100]}...')
    print()
```
